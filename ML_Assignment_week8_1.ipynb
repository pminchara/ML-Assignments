```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["figure.figsize"]=[12,9]
plt.rcParams["axes.grid"] = False
```

```
def conv2d(input,kernel):
  m, n = kernel.shape
  x, y = input.shape
  x = x - m + 1
  y = y - n + 1
  conv = np.zeros((x,y))
  for i in range(x):
    for j in range(y):
      conv[i][j] = np.sum(kernel*(input[i:i+m, j:j+n]))
  plt.imshow(conv)
  return conv
```
