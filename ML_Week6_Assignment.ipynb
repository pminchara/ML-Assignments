```
def ridge(c,g,X_train,y_train,X_test,plot):
  model = KernelRidge(alpha=1.0/(2*c), kernel='rbf', gamma=g).fit(X_train, y_train)
  print("The dual coefficients \n",model.dual_coef_)
  y_pred = model.predict(X_test)
  if(plot):
    plt.scatter(X_train, y_train, color='red', marker='+')
    plt.plot(X_test, y_pred, color='blue')
    plt.xlabel("input x"); plt.ylabel("output y")
    plt.legend(["Kernel Ridge","train"])
    plt.show()
  return model,y_pred
```

```
mean_error_ridge=[]
std_error_ridge=[]
Ci_range = [0.1, 1, 10, 100 ,1000]
for Ci in Ci_range:
  model_ridge = KernelRidge(alpha=1.0/(2*Ci), kernel='rbf', gamma=1)
  temp_ridge=[]
  for train, test in kf.split(X):
    model_ridge.fit(X[train], y[train])
    ypred_ridge = model_ridge.predict(X[test])
    temp_ridge.append(mean_squared_error(y[test],ypred_ridge))
  mean_error_ridge.append(np.array(temp_ridge).mean())
  std_error_ridge.append(np.array(temp_ridge).std())
plt.errorbar(Ci_range,mean_error_ridge,yerr=std_error_ridge)
plt.xlabel('Ci')
plt.ylabel('Mean square error')
plt.show()
```

```
mean_error_ridge_gamma=[]
std_error_ridge_gamma=[]
gamma_range = [0,1,5,10,25]
for g in gamma_range:
  model_ridge_gamma = KernelRidge(alpha=1.0/(2*100), kernel='rbf', gamma=g)
  temp_ridge_gamma=[]
  for train, test in kf.split(X):
    model_ridge_gamma.fit(X[train], y[train])
    ypred_ridge_gamma = model_ridge_gamma.predict(X[test])
    temp_ridge_gamma.append(mean_squared_error(y[test],ypred_ridge_gamma))
  mean_error_ridge_gamma.append(np.array(temp_ridge_gamma).mean())
  std_error_ridge_gamma.append(np.array(temp_ridge_gamma).std())
plt.errorbar(gamma_range,mean_error_ridge_gamma,yerr=std_error_ridge_gamma)
plt.xlabel('gamma')
plt.ylabel('Mean square error')
plt.show()
```
